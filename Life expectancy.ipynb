There were certain extreme values for BMI.After checking the web for the minimum BMI value, I replaced the extreme minimum values with 12

```python
for i in data[' BMI ']:
    if i <12:
        data[' BMI '].replace(i,12, inplace=True)
#data[' BMI ']=data[' BMI '].mask[data[' BMI ']<12,12]
```

I replaced the mull values in the target variable,life expectancy with the median value of the column

```python
data['Life expectancy ']=data['Life expectancy '].replace(np.nan,72.1)
```

I encoded the categorical columns Country and Status using the get dummies method

```python
status = pd.get_dummies(data['Status'],drop_first=True)
data.drop(['Status'],axis=1,inplace=True)
data = pd.concat([data,status],axis=1)
```

```python
country = pd.get_dummies(data['Country'],drop_first=True)

data.drop(['Country'],axis=1,inplace=True)
data = pd.concat([data,country],axis=1)
```
